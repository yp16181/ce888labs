```python
import sys
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
#This program just finish the target that convert tasks words to fixed length representations 
#of vectors by GloVe arithmetic and feed them to a Random Forest

#build the dictionary for model feature X and target Y
glove_words = []
glove_vectors = []
filename = "vectors.txt"


with open(filename) as f:
    for line in f:
        l = line.strip().split(' ')
        glove_words.append(l[0])
        glove_vectors.append(np.asarray(map(np.float32, l[1:]), dtype='float'))
glove = dict(zip(glove_words, glove_vectors))

print (glove_words)
print (glove_vectors)
```

```python
#set up the model feature X
X = np.array([glove_words])
#set up the target Y
Y = np.array([glove_vectors])


#set up the RandomForest model
clf = RandomForestClassifier(n_estimators=20, max_depth=5)


#Train the RandomForest model by vecters from GloVe
clf.fit(glove_vectors,glove_words)
```

```python
#Test the RandomForest model
clf.score(glove_vectors, glove_words)

```
